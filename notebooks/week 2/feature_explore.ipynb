```python
# calculate interquartile range to detect outliers

def IQR(df, column, up_q=0.8, low_q=0.2, kordaja=1.5):
    Q1 = df[column].quantile(low_q)
    Q2 = df[column].quantile(up_q)

    iqr = Q2 - Q1

    lower_bound = Q1 - kordaja * iqr
    upper_bound = Q2 + kordaja * iqr

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers
```
